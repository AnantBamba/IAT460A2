```python
import ColabTurtle.Turtle as t
from IPython.display import clear_output
import random
import turtle as turtle
```

Creating a floral garden

```python
def create_l_system(iterations, axiom, rules):
    """Generate L-System instructions based on axiom and rules."""
    result = axiom
    for _ in range(iterations):
        new_string = ""
        for char in result:
            new_string += rules.get(char, char)
        result = new_string
    return result

def draw_l_system(instructions, angle, distance, anglepetal, num_plants):
    """Draw the L-System using turtle graphics.

    Parameters:
    - instructions: string of L-System commands
    - angle: turning angle in degrees
    - distance: forward movement distance
    - anglepetal: turning angle for petals
    - num_plants: number of plants to draw
    """
    stack = []
    for _ in range(num_plants): #for number of plants to be drawn
        x_pos = random.randint(50, 800)  # Random x position
        y_pos = random.randint(100, 600)   # Random y position
        t.penup()
        t.goto(x_pos, y_pos)
        t.pendown()
        for cmd in instructions:
            if cmd == 'F':  # Move forward and draw
                t.pensize(2)
                t.pencolor("green")
                t.forward(distance)
            elif cmd == 'P':  # petal colour
                t.pensize(3)
                t.pencolor("pink")
                t.forward(distance)
            elif cmd == 'T':  # petal colour
                t.pensize(3)
                t.pencolor("white")
                t.forward(distance)
            elif cmd == 'S':  # spike colour
                t.pencolor("red")
                t.forward(distance)
            elif cmd == 'f':  # Move forward without drawing
                t.penup()
                t.forward(distance)
                t.pendown()
            elif cmd == '+':  # Turn right
                t.right(angle)
            elif cmd == '-':  # Turn left
                t.left(angle)
            elif cmd == 'b':  # Turn left for petal
                t.left(anglepetal)
            elif cmd == 'd':  # Turn right for petal
                t.right(anglepetal)
            elif cmd == '[':  # Save current state
                stack.append((t.position(), t.heading()))
            elif cmd == ']':  # Restore previous state

                position, heading = stack.pop()
                t.penup()
                t.goto(position)
                t.setheading(heading)
                t.pendown()

def setup_turtle():
        t.initializeTurtle()
        t.hideturtle()
        t.speed(13)  # Fastest speed
        t.penup()
        t.goto(t.window_width() // 2, t.window_height() - 50)  # Start position
        t.pendown()
```

```python
# Plant parameters
plant_axiom = "X"
Flytrap_rules = {
    "S": "FbFbFbFbF",
    "X": "[F[][-X]+X]+S+S",
    "F": "F"
}

Lav_rules={
    "X": "F[-X]F[X]F+FP",
    "P": "P+P"
}

Dandy_rules = {

     "X": "FF+[FT]",
    "F": "F",
    "T": "TbT",
}
#parameters
plant_iterations_Flytrap = 6
plant_iterations_lav=4
plant_iterations_a=6
plant_angle_l = 10
petal_angle_l = 20
plant_angle_f = 25
petal_angle_f = 90
plant_angle_a = 60
petal_angle_a = 50
numberl = 6
numberf= 5
#Adding all thre planst together
# Generate and draw
setup_turtle()
t.pensize(1)
lav_instructions = create_l_system(plant_iterations_lav, plant_axiom, Lav_rules)
draw_l_system(lav_instructions, plant_angle_l, 10,petal_angle_l,numberl)
ft_instructions = create_l_system(plant_iterations_Flytrap, plant_axiom, Flytrap_rules)
draw_l_system(ft_instructions, plant_angle_f, 10,petal_angle_f,numberf)
ft_instructions = create_l_system(plant_iterations_Flytrap, plant_axiom, Dandy_rules)
draw_l_system(ft_instructions, plant_angle_a, 10,petal_angle_a,20)
```

```python
ft_instructions = create_l_system(plant_iterations_a, plant_axiom, Dandy_rules)
draw_l_system(ft_instructions, plant_angle_a, 10,petal_angle_a,20) #Adding more dandelions
```
